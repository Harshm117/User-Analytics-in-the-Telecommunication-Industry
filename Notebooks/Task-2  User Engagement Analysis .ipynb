```python
data = telcom.copy()
data['Total Traffic (Bytes)']= data['Total DL (Bytes)'] + data["Total UL (Bytes)"]
data["Total Youtube Volume(Bytes)"]= data["Youtube DL (Bytes)"] + data["Youtube UL (Bytes)"]
data["Total Google Volume(Bytes)"]= data["Google DL (Bytes)"] + data["Google UL (Bytes)"]
data["Total Email Volume(Bytes)"]= data["Email DL (Bytes)"] + data["Email UL (Bytes)"]
data["Total Social Media Volume(Bytes)"]= data["Social Media DL (Bytes)"] + data["Social Media UL (Bytes)"]
data["Total Netflix Volume(Bytes)"]= data["Netflix DL (Bytes)"] + data["Netflix UL (Bytes)"]
data["Total Gaming Volume(Bytes)"]= data["Gaming DL (Bytes)"] + data["Gaming UL (Bytes)"]
data["Total Other Volume(Bytes)"]= data["Other DL (Bytes)"] + data["Other UL (Bytes)"]
```

```python
## Aggregate total data volume (in Bytes) during this session for each application
apps_data_volume = data.groupby(['Dur. (ms)','IMSI', 'MSISDN/Number']).agg({
    'Total Traffic (Bytes)': 'sum',
    'Total Youtube Volume(Bytes)': 'sum',
    'Total Google Volume(Bytes)': 'sum',
    'Total Email Volume(Bytes)': 'sum',
    'Total Social Media Volume(Bytes)': 'sum',
    'Total Netflix Volume(Bytes)': 'sum',
    'Total Gaming Volume(Bytes)': 'sum',
    'Total Other Volume(Bytes)': 'sum',
}).reset_index()

apps_data_volume
```

```python
# Refactored code
engagement_metrics = data.groupby('MSISDN/Number').agg({'Bearer Id': 'count', 'Dur. (ms)': 'sum', 'Total Traffic (Bytes)': 'sum'})
engagement_metrics = engagement_metrics.rename(columns={'Bearer Id': 'Sessions Frequency', 'Dur. (ms)': 'Sessions Duration (ms)'})

engagement_metrics = engagement_metrics.sort_values(by=['Sessions Frequency', 'Sessions Duration (ms)', 'Total Traffic (Bytes)'], ascending=False).head(10)
engagement_metrics
```

## Using the k-means clustering algorithm, group users in k engagement clusters based on the engagement metrics: 
- What is the optimized value of k (use the elbow method for this)?  
- Interpret your findings. 


```python
# the user engagement metrics df (already normalized due to different units of measurement)
norm_engagement_metrics = norm_engagement_metrics.drop(['cluster'], axis = 1)
norm_engagement_metrics.head()
```

```python
app_user.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm


# Extract engagement metrics
engagement_metrics = app_user[['Total Youtube Volume(Bytes)', 
                                       'Total Google Volume(Bytes)', 
                                       'Total Email Volume(Bytes)', 
                                       'Total Social Media Volume(Bytes)', 
                                       'Total Netflix Volume(Bytes)', 
                                       'Total Gaming Volume(Bytes)', 
                                       'Total Other Volume(Bytes)']]

# Standardize the data
scaler = StandardScaler()
engagement_metrics_scaled = scaler.fit_transform(norm_engagement_metrics)

# Use the elbow method to find the optimal number of clusters
sse = []
for k in tqdm(range(1, 11)):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(engagement_metrics_scaled)
    sse.append(kmeans.inertia_)

# Plot the SSE for each value of k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('The Elbow Method for Determining the Number of Clusters')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances (SSE)')
plt.xticks(range(1, 11))
plt.tight_layout()
plt.show()
```

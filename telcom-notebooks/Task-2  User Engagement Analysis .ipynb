```python
#data = telcom.copy()
telcom['Total Traffic (Bytes)']= telcom['Total DL (Bytes)'] + telcom["Total UL (Bytes)"]
telcom["Total Youtube Volume(Bytes)"]= telcom["Youtube DL (Bytes)"] + telcom["Youtube UL (Bytes)"]
telcom["Total Google Volume(Bytes)"]= telcom["Google DL (Bytes)"] + telcom["Google UL (Bytes)"]
telcom["Total Email Volume(Bytes)"]= telcom["Email DL (Bytes)"] + telcom["Email UL (Bytes)"]
telcom["Total Social Media Volume(Bytes)"]= telcom["Social Media DL (Bytes)"] + telcom["Social Media UL (Bytes)"]
telcom["Total Netflix Volume(Bytes)"]= telcom["Netflix DL (Bytes)"] + telcom["Netflix UL (Bytes)"]
telcom["Total Gaming Volume(Bytes)"]= telcom["Gaming DL (Bytes)"] + telcom["Gaming UL (Bytes)"]
telcom["Total Other Volume(Bytes)"]= telcom["Other DL (Bytes)"] + telcom["Other UL (Bytes)"]
```

```python
## Aggregate total data volume (in Bytes) during this session for each application
apps_data_volume = telcom.groupby(['Dur. (ms)','IMSI', 'MSISDN/Number']).agg({
    'Total Traffic (Bytes)': 'sum',
    'Total Youtube Volume(Bytes)': 'sum',
    'Total Google Volume(Bytes)': 'sum',
    'Total Email Volume(Bytes)': 'sum',
    'Total Social Media Volume(Bytes)': 'sum',
    'Total Netflix Volume(Bytes)': 'sum',
    'Total Gaming Volume(Bytes)': 'sum',
    'Total Other Volume(Bytes)': 'sum',
}).reset_index()

apps_data_volume.head()
```

### Sessions frequency

```python
# Calculate the sessions frequency for each user
sessions_frequency = telcom['MSISDN/Number'].value_counts()

sessions_frequency = pd.DataFrame({'Sessions Frequency': sessions_frequency}).reset_index()

# Display the head of the sessions frequency series
sessions_frequency.head()
```

***The table above shows the top five users with the highest session frequency in the dataset. The session frequency is the count of sessions per user, identified by their MSISDN/Number. The user with the highest number of sessions has 1065 sessions, followed by one user with 18 sessions, and so on.***

```python
## Plot the distribution of sessions frequency
plt.figure(figsize=(10, 6))
sessions_frequency.hist(bins=50)
plt.title('Distribution of Total Session Frequency per User')
plt.xlabel('Total Frequency')
plt.ylabel('Number of Users')
plt.yscale('log')
plt.grid(True)
plt.show()
```

## The Duration of the session

```python
# Convert duration from milliseconds to seconds for better readability
telcom['Dur. (s)'] = telcom['Dur. (ms)'] / 1000

## Calculate the duration of each session
session_duration = telcom.groupby('MSISDN/Number')['Dur. (ms)'].sum()

## Convert the result into a DataFrame
session_duration = pd.DataFrame({'Session Duration (ms)': session_duration}).reset_index()

## Display the head of the session durations series
session_duration.head(10)
```

***The table above shows the total duration of sessions for the top five users in the dataset, measured in milliseconds. Each row represents a unique user identified by their MSISDN/Number and the corresponding sum of durations for all sessions they had.***

```python
plt.figure(figsize=(10, 6))
sessions_frequency.hist(bins=50) #range=(1, sessions_frequency.max()))
plt.title('Distribution of Session Frequency per User')
plt.xlabel('Number of Sessions')
plt.ylabel('Number of Users')
plt.yscale('log')
plt.grid(True)
plt.show()
```

## The session total traffic (download and upload (bytes))


```python
## Calculate the total traffic for each session by summing up the download and upload bytes
telcom['Total Traffic (Bytes)'] = telcom['Total DL (Bytes)'] + telcom['Total UL (Bytes)']

## Group by 'MSISDN/Number' to find the total traffic for each user
session_total_traffic = telcom.groupby('MSISDN/Number')[['Total Traffic (Bytes)']].sum().reset_index()



## Display the head of the session total traffic dataframe
session_total_traffic.head()
```

```python
## Plot the distribution of session total traffic
plt.figure(figsize=(10, 6))
session_total_traffic.hist(bins=50)
plt.title('Distribution of Total Traffic per User')
plt.xlabel('Total Traffic (Bytes)')
plt.ylabel('Number of Users')
plt.yscale('log')
plt.grid(True)
plt.show()
```

***The table displays the total download and upload traffic for the top five users, along with their combined total traffic.***

```python
## Top 10 customers per engagement metric
engagement_metrics = telcom.groupby('MSISDN/Number').agg({'Bearer Id': 'count', 'Dur. (ms)': 'sum', 'Total Traffic (Bytes)': 'sum'})
engagement_metrics = engagement_metrics.rename(columns={'Bearer Id': 'Sessions Frequency', 'Dur. (ms)': 'Session Duration (ms)'})

engagement_metrics = engagement_metrics.sort_values(by=['Sessions Frequency', 'Session Duration (ms)', 'Total Traffic (Bytes)'], ascending=False).head(10)
engagement_metrics.reset_index()
```

### The top 10 customers have been identified for each engagement metric:

- Session frequency: The number of sessions each customer had.
- Session duration: The total duration of all sessions per customer.
- Total traffic: The total data (download and upload) consumed per customer.

## Normalize each engagement metric and run a k-means (k=3) to classify customers into three groups of engagement.

```python
# Normalizing the engagement_metrics using log transformation
engagement_metrics_normalized = np.log1p(engagement_metrics)
engagement_metrics_normalized.head()
```

```python
engagement_metrics_normalized.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
# kmeans on 3 clusters.
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
kmeans.fit(engagement_metrics_normalized)
kmeans_data = kmeans.predict(engagement_metrics_normalized)
```

```python
# add the predicted clusters as a column in the df
engagement_metrics_normalized['Cluster'] = pd.Series(kmeans_data, index=engagement_metrics_normalized.index)
engagement_metrics_normalized.head().reset_index()
```

***The scatter plot visualizes the customer engagement clusters based on normalized session duration and total traffic. Each point represents a customer, with the color indicating the cluster to which they have been assigned by the k-means algorithm.***

```python
engagement_metrics_normalized['Cluster'].value_counts()
```

```python
#plot the clusters(3D)
pd.plotting.parallel_coordinates(engagement_metrics_normalized, 'Cluster')
```

```python
# determining number of clusters using elbow plot
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(engagement_metrics_normalized)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```
